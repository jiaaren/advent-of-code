```python
with open('input.txt') as f:
    file = f.read()
    lines = file.split('\n')

test_calc_sizes = 0

def calc_sizes(node):
    # accidentally included this earlier
    # if node.files == []:
    #     return 0

    # calc total file size of all nodes below
    child_sizes = [calc_sizes(child) for child in node.dirs]

    current_file_size = node.get_dir_size()

    global test_calc_sizes
    test_calc_sizes += current_file_size
    return current_file_size + sum(child_sizes)

class Node:
    def __init__(self, name, parent, dirs: list = [], files: list = []):
        self.name = name
        self.dirs = dirs
        self.files = files
        self.nextval = None
        self.parent = parent # Node

    def get_node(self, node_name):
        for node in self.dirs:
            if node.name == node_name:
                return node

    def get_dir_size(self, rec=True):
        sizes = [int(x[0]) for x in self.files]
        total_sum = sum(sizes)
        return total_sum


root = Node('root', None, [], [])
loc = root

is_ls = False

total_file_size = 0

for l in lines:
    print(l + ' - ', end='')
    if l == '$ cd /':
        loc = root
    if l[0] == '$':
        is_ls = False

    if is_ls:
        f, s = l.split()
        if f == 'dir':
            new_node = Node(s, loc, [], [])
            loc.dirs.append(new_node)

            print(f'make new dir {s}')
        else:
            loc.files.append((f, s))
            total_file_size += int(f)
            print(f'make new file {s}')
    else:
        if l == '$ ls':
            is_ls = True
            print('check files')
        elif '$ cd ' in l :
            dir_name = l.split()[-1]
            if dir_name == '..':
                loc = loc.parent
                print(f'go back now in parent {loc.name}')
            elif dir_name != '/':
                addr = loc.get_node(dir_name)
                loc = addr
                print(f'go in now in child {loc.name}')
                # print(addr)
```

```python
results = []
below_100000 = 0

def gen_res1(node):
    total_dir_size = calc_sizes(node)
    results.append((node.name, total_dir_size))

    if total_dir_size <= 100000:
        global below_100000
        below_100000 += total_dir_size

    for child in node.dirs:
        gen_res1(child)

gen_res1(root)
```

```python
import pandas as pd 
pd.DataFrame(results).to_csv('res.csv')
```
