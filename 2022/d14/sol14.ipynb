```python
with open('input.txt') as f:
    file = f.read()

rocks_mvmt = []
min_x = min_y = 99999999999999 
max_x = max_y = 0
for l in file.split('\n'):
    tmp = []
    pos_lst = l.split(' -> ')
    for pos in pos_lst:
        x, y = pos.strip().split(',')
        y = int(y)
        x = int(x)
        tmp.append((x, y))

        # ternary for if - https://stackoverflow.com/questions/12199757/python-ternary-operator-without-else
        if x > max_x: max_x = x
        if y > max_y: max_y = y
        if x < min_x: min_x = x
        if y < min_y: min_y = y

    rocks_mvmt.append(tmp)

# + 3 for x axis to add border on both sides, + 2 for y axis to add border on bottom only
grid = [['.' for x in range(max_x - min_x + 3)] for y in range(0, max_y + 2)]
for i in range(len(grid)):
    grid[i][0] = '!'
    grid[i][-1] = '!'
for i in range(len(grid[0])):
    grid[-1][i] = '!'
```

```python
def print_grid(grid=grid):
    for row in grid:
        print(' '.join(row))

def scale_mvmt(mvmt):
    # no need to scale y
    # print(mvmt[0] - min_x, mvmt[1] - min_y)
    return mvmt[0] - min_x + 1, mvmt[1] + 1

def update_mvmt(mvmt_lst, scale_mvmt=scale_mvmt, grid=grid):
    for i, mvmt in enumerate(mvmt_lst):
        x, y = scale_mvmt(mvmt)
        grid[y][x] = '#'

        if i != 0:
            prev_x, prev_y = scale_mvmt(mvmt_lst[i - 1])
            if prev_x == x:
                high_y, low_y = max(prev_y, y), min(prev_y, y)
                for a in range(low_y, high_y):
                    grid[a][x] = '#'
            elif prev_y == y:
                high_x, low_x = max(prev_x, x), min(prev_x, x)
                for b in range(low_x, high_x):
                    grid[y][b] = '#'

def update_mvmt_all(rocks_mvmt=rocks_mvmt, scale_mvmt=scale_mvmt, grid=grid):
    for mvmt_lst in rocks_mvmt:
        update_mvmt(mvmt_lst, scale_mvmt=scale_mvmt, grid=grid)
        
update_mvmt_all()
# print_grid()
```

```python
def trickle_one(start=(500, 0), scale_mvmt=scale_mvmt, grid=grid):
    x, y = scale_mvmt(start)
    prev_x = prev_y = 0

    while True:
        # keep track whether any movements since previous trickle
        prev_x, prev_y = x, y
        # try to go down
        if grid[y+1][x] in ['.', '!']:
            y += 1
        # try go left
        elif grid[y+1][x-1] in ['.', '!']:
            y += 1
            x -= 1
        # try go right
        elif grid[y+1][x+1] in ['.', '!']:
            y += 1
            x += 1
        if grid[y][x] == '!':
            return False
        if prev_x == x and prev_y == y:
            grid[y][x] = 'o'
            return True
```

```python
i = 0
while True:
    no_leak = trickle_one()
    if no_leak == False:
        break
    i += 1
    # print_grid()
print(i)
```

```python
def scale_mvmt2(mvmt):
    # no need to scale y
    # print(mvmt[0] - min_x, mvmt[1] - min_y)
    return mvmt[0] - min_x + len(grid[0]), mvmt[1] + 1

update_mvmt_all(scale_mvmt=scale_mvmt2, grid=grid2)
print_grid(grid2)
```

```python
trickle_x, trickle_y = scale_mvmt2((500, 0))
i = 0
while True:
    no_leak = trickle_one(scale_mvmt=scale_mvmt2, grid=grid2)
    if no_leak == False:
        print('Leaking!')
        break
    i += 1
    if grid2[trickle_y][trickle_x] == 'o':
        print('Hit trickle!')
        break
    # print_grid()
print(i)
```
