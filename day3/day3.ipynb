```python
import pandas as pd
import numpy as np
day3 = pd.read_csv('day3.csv', dtype={'input': np.str})
```

```python
day3
```

```python
binary = day3['input'].str.split(pat='', expand=True)
binary = binary.drop([0, 13], axis=1)
binary
```

```python
binary = binary.astype(int)
binary
```

```python
res = binary.sum(axis=0)
res
```

```python
res > 500
```

```python
gamma = ""
epsilon = ""

for i in (res > 500):
    gamma += str(int(i))
    # return inverse of boolean with not
    # https://stackoverflow.com/questions/7030831/how-do-i-get-the-opposite-negation-of-a-boolean-in-python
    epsilon += str(int(not i))

print(f'gamma binary - {gamma}')
print(f'gamma int - {int(gamma, 2)}')
print(f'epsilon binary - {epsilon}')
print(f'epsilon int - {int(epsilon, 2)}')
```
